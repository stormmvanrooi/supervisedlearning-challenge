```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
del train_df['Unnamed: 0']
del train_df['index']
del test_df['Unnamed: 0']
del test_df['index']
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
# Convert categorical data to numeric and separate target feature for training data

train_df = pd.get_dummies(train_df, columns=['home_ownership','verification_status',
                                             'application_type','hardship_flag',
                                             'debt_settlement_flag','initial_list_status',
                                             'pymnt_plan','loan_status'])
```

```python
# Convert categorical data to numeric and separate target feature for testing data
# add missing dummy variables to testing set
test_df = pd.get_dummies(test_df, columns=['home_ownership','verification_status',
                                           'application_type','hardship_flag',
                                           'debt_settlement_flag','initial_list_status',
                                           'pymnt_plan','loan_status'])
```

```python
del train_df['loan_status_low_risk']
del test_df['loan_status_low_risk']
```

```python
train_df=train_df.rename(columns={'loan_status_high_risk':'loan_status'})
```

```python
test_df =test_df.rename(columns={'loan_status_high_risk':'loan_status'})
```

```python
test_df['debt_settlement_flag_Y'] = 0
```

```python
y_train = train_df['loan_status']
X_train = train_df.drop('loan_status', axis=1)
```

```python
y_test = test_df['loan_status']
X_test = test_df.drop('loan_status', axis=1)
```

# Model Prediction
#### The random forest model will be a better predictor of credit score risk because it is better at determining the class that a value belongs to. Due to the nature of the data, the random forest model will be more accurate. 

```python
# Train the Logistic Regression model on the unscaled data and print the model score
unscaled_logreg = LogisticRegression()
unscaled_logreg.fit(X_train, y_train)
```

```python
# Train a Random Forest Classifier model and print the model score
unscaled_ranfor = RandomForestClassifier(random_state=1, n_estimators=3).fit(X_train, y_train)
```

```python
print(f"Testing Data Score for unscaled LR: {unscaled_logreg.score(X_test, y_test)}")
print(f'Testing Score for unscaled RF: {unscaled_ranfor.score(X_test, y_test)}')
```

```python
# Scale the data
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
scaled_logreg = LogisticRegression().fit(X_train_scaled, y_train)
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
scaled_ranfor = RandomForestClassifier(random_state=1, n_estimators=3).fit(X_train_scaled, y_train)
```

```python
print(f'Testing Score for scaled LR: {scaled_logreg.score(X_test_scaled, y_test)}')
print(f'Testing Score for scaled RF: {scaled_ranfor.score(X_test_scaled, y_test)}')
```

# Model Evaluation
#### The logistic regression model is the better fit of the two machine learning models. When scaled, the accuract of the logistic regression model improved. The random forest model accuracy remained the same when the data was unscaled and scaled.
